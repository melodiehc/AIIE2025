Python has various "types" of data. In this notebook we will focus on following numeric data:

 integers and floating point numbers.

1.   Integers a.k.a 'int'
2.   Floating Point a.k.a. 'float'



Integers are whole numbers, positive or negative. For example: 2 and -2 are examples of integers.

The size of Pythonâ€™s integers is limited only by machine memory, not by a fixed number of bytes.

Floating point numbers in Python are notable because they have a decimal point in them, or use an exponential (e) to define the number. For example 2.0 and -2.1 are examples of floating point numbers. 4E2 (4 times 10 to the power of 2) is also an example of a floating point number in Python.

We can use a <code>:</code> to perform *slicing* which grabs everything up to a designated point. For example:

```python
# Grab everything PAST the first term all the way to the length of s which is len(s)
s[1:]
```

```python
# Grab everything UP TO the 3rd index
s[:3]
```

```python
# Grab everything, but go in step sizes of 2
s[::2]
```

```python
# Call values by their key
my_dict1['key2']
```

```python
#Dictionary with different Python objects
my_dict2 = {'key1':123,'key2':[12,23,33],'key3':['item0','item1','item2']}
```

### Accessing the items from a Dictionary

```python
# Call items from the dictionary
my_dict2 = {'key1':123,'key2':[12,23,33],'key3':['item0','item1','item2']}
my_dict2['key3']
```

```python
#To get the first item of value which is a list object
my_dict2['key3'][0]
```

```python
my_dict2['key1'] = my_dict2['key1']*10
```
