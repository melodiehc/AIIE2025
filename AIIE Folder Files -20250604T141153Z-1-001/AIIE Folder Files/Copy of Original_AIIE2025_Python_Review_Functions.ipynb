### *args and **kwargs

```python
def myfunc(a,b):
    return sum((a,b))

myfunc(40,60)
```

In this example, a and b are positional arguments; that is, 40 is assigned to **a** because it is the first argument, and 60 to **b**.

What if we want to work with more than two numbers?

    * One way would be to assign a lot of parameters and give each
      one a default value.
    * Use  *args

```python
def myfunc(a=0,b=0,c=0,d=0,e=0):
    return sum((a,b,c,d,e))

myfunc(40,60,20)
```

```python
def myfunc(*args):
    return sum(args)
```

```python
myfunc(40,60,20)
```

```python
myfunc(40,60,20,50)
```

```python
myfunc(40,60,20,30,50, 10)
```

 #### **kwargs

```python
def myfunc(**kwargs):
  if 'animal' in kwargs:
    print("My favorate animal is {}".format(kwargs['animal']))
  elif 'car' in kwargs:
    print("My favorate car is {}".format(kwargs['car']))
  else:
    print("Not found")
```

```python
myfunc(car='Tesla')
```

```python
def myfunc(*args, **kwargs):
  print(args)
  print(kwargs)
  print("I want {} {}".format(args[0],kwargs['car']))
```

```python
myfunc(10,50, car='Tesla', food="salad", animal='cat')
```
