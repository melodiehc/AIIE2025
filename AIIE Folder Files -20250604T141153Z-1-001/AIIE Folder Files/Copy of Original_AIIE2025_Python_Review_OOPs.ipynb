Abstraction
    
    Abstraction is a process of hiding the implementation details from the user, only the functionality will be provided to
    the user. In other words, the user will have the information on what the object does instead of how it does it.

    You need to know how to use the phone you do not need to know how to built it.


Encapsulation
    
    Encapsulation is a mechanism where you bind your fields/variables and code together as a single unit. It also
    means to hide your data in order to make it safe from any modification.


Inheritance

    An Inheritance is basically a way of forming new classes using the classes that have already been defined.
    
    The significant advantages of Inheritance is the ability to reuse the code that we have already worked on and reduce the
    complexity of the program.

    The existing(already defined) class is called the base class and the new class is known as the derived class since it is reusing
    the code of base class

Polymorphism

#Inheritance

```python
# Creating Base Class
class Animal():
  def __init__(self):
    print("Animal Created")

  def who_am_i(self):
    print("I am an Animal")

  def eat(self):
    print("I am eating")
```

```python
# Derived Class that is Inheriting the code of Base Class
# Dog inherits Animal
class Dog(Animal):
  def __init__(self):
    # Inherit the instance of Animal Class when creating the instance of Dog.
    Animal.__init__(self)
    print("Dog Created")
    # Overwriting the Base Class Method
  def who_am_i(self):
    print("I am a Dog")
    # Add on Method
  def bark(self):
    print("Woof!")
```

```python
my_dog = Dog()
```

```python
my_dog.who_am_i()
```

#Polymorphism

```python
# Mention Pdf file_type class that has method file_open
class Pdf():
  def __init__(self,name):
    self.name = name
  # Call the filename and print the pdf
  def file_open(self):
    return self.name + " Document opened"
```

```python
# Creating similar file_type class for Excel with file_open
class Excel():
  def __init__(self,name):
    self.name = name
  # Call the filename and print the excel
  def file_open(self):
    return self.name + " Sheet opened"
```

```python
# Instance of class Pdf with filename Invoice
Invoice = Pdf("Invoice")

# Instance of class Excel with filename Revenue
Revenue = Excel("Revenue")
```

```python
print(Invoice.file_open())
```

```python
print(Revenue.file_open())
```

```python
# Iterate through list of files
for file_type in [Invoice, Revenue]:
  # print type of file
  print(type(file_type))
  # print the result of method file_open
  print(file_type.file_open())
```
